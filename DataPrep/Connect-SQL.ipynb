# Azure SQL Database との連携

##使用するデータのアップロード

ローカル環境にある diabetes.csv を左の Data タブよりアップロードします。登録された Table の名前は diabetes_csv の前提で下記コードを進めます。

```
df = spark.table("diabetes_csv")
```

```
#display(df)
```

## JDBC接続情報の設定

```
jdbcHostname = ""
jdbcDatabase = ""
jdbcPort = 1433
username =""
password = ""
jdbcUrl = "jdbc:sqlserver://{0}:{1};database={2};user={3};password={4}".format(jdbcHostname, jdbcPort, jdbcDatabase, username, password)
print(jdbcUrl)
```

## データ加工 (列追加)

```
df = df.withColumn("age/100", df.AGE/100)
```

```
#display(df)
```

## データの書き込み

```
# mode : "append", "overwrite", "ignore", "error"
df.write.jdbc(table="dbo.diabetes_out", url=jdbcUrl, mode= "overwrite")
```

## データの取得

```
df2 = spark.read.jdbc(url=jdbcUrl, table="diabetes_out", lowerBound=0, upperBound=100000, numPartitions=100)
```

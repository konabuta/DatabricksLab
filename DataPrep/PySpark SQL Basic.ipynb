# PySpark & Spark SQL 基礎操作

## DataFrameの作成

```
# Workshopで使用するデータの確認
dbutils.fs.ls("/mnt/lab")
```

```
# データのロード
adult = spark.read.options(header='true', inferSchema='true').csv('/mnt/lab/adult.csv')
```

```
# データ表示
display(adult)
```

```
# スキーマ情報
adult.printSchema()
```

## DataFrame 基本操作

```
# 列の選択 (単一)
adult.select("workclass").show(5)
#adult.select("workclass").distinct().show(5)
```

```
# 列の選択 (複数)
adult.select("age","education_num").show(5)
```

```
# 基礎統計量
display(adult.describe())
```

```
# カウント
adult.count()
```

```
# フィルタリング
display(adult.filter(adult["age"] > 17.0))
```

```
# 新規計算列
adult = adult.withColumn("education_day", adult.education_num * 365)
display(adult)
```

```
# 列の削除
adult = adult.drop("education_day")
display(adult)
```

```
# グラフ表示
display(adult)
```

## SQL Table 登録

```
# Tmp Tableとして登録
adult.createOrReplaceTempView('adultTB')
```

```
 %sql
 select * from adultTB
```

```
%sql
select distinct age from adultTB order by age
```

```
tableDF = spark.sql("select distinct age from adultTB order by age")
display(tableDF)
```
